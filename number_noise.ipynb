```python
import re

def clean_text(text):
    # 1. 숫자가 문자 사이에 있는 경우 숫자를 공백으로 대체
    text = re.sub(r'(?<=[a-zA-Z가-힣])\d+(?=[a-zA-Z가-힣])', ' ', text)

    # 2. 특수 기호가 '.'나 '…', '~'가 아닌 경우에 한해 숫자를 제거
    text = re.sub(r'(?<![.\u2026∼])([^\w\s.\u2026∼])\d+', r'\1', text)

    # 불필요한 공백 정리
    cleaned_text = re.sub(r'\s+', ' ', text).strip()
    
    return cleaned_text

# 테스트
#text = "이 문장에는 yo39종, !100, #50, 그리고 200명, 3.14와 7일 33㎡ 35% 강수량 50㎜이 있습니다."
text = "STJ레콤 J분기 영D익t4천105t…2∼3월 1∼2일 5명 100명 3.14 2>1％"
result = clean_text(text)
print(result)
```

```python
df['text'] = df['text'].apply(clean_text)
```
